#### 6. Use this variable to create two variables that are called "first_position" and "second_position". Hint: If a player plays two positions, then the "-" is used as a delimiter between their first and second positions. In these cases, you will have to split the position variable into two. <a class="anchor" id="bullet6"></a>

```python
df[['first_position', 'second_position']] = df.Position.str.split("-", expand = True).fillna("")

df
```

#### 7. Combine these two variables back into a single variable called "position_united". Make sure to separate the two positions by "-", as the original data field did. If there is no second position, then there will only be a single position. Again, this is similar to how the original variable was coded. Present a frequency table of "position_united".<a class="anchor" id="bullet7"></a>

```python
df['position_united'] = df.apply(lambda a: '-'.join(filter(None, [a['first_position'], a['second_position']])), axis=1)
df
```

#### 10. Subset the original dataset to the years 1995-1997. Group the data by team name and year and then summarize the average number of points per team. Arrange from most to least points.<a class="anchor" id="bullet10"></a>

#### 11. Similar to Q10, start by subsetting the original dataset to the years 1995-1997. Then create a wide dataset that keeps year in a single row, but spreads team names to multiple columns with each column delineating points per team in 1995-1997 seasons (Hint: Your final answer should have a shape of 3 x 30).. <a class="anchor" id="bullet11"></a>

```python
df = pd.read_csv(path + file)
df = df[(df["Year"] >= 1995) & (df["Year"] <= 1997)]
df1 = df.groupby(["Tm", "Year"])["PTS"].sum().reset_index()

pivoted = df1.pivot(index="Year", columns="Tm", values="PTS") # Switch back to original / wide format
pivoted
```

#### 12.   Now return the data from #11 back into a long (tidy) format by moving Team (TM), Points and Year into their own columns (Hint: Your data shape should be 90 x 3)  <a class="anchor" id="bullet12"></a>

```python
long_data = (pivoted.stack()   # use .stack() method for converting to long
              .reset_index()
              .rename(columns={0: "PTS"}))

long_data
```
